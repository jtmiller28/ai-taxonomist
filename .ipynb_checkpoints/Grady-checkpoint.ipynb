```python
import csv
import os
from pyinaturalist import get_observations
import requests

# Step 1: Search for observations of the species Calycoseris wrightii, specifying per_page=200 to get more results per page
observations = get_observations(taxon_name='Calycoseris wrightii', per_page=200)

# Define the CSV file path
csv_file = 'Cw_image_metadata.csv'

# Ensure the output folder exists
output_folder = 'Cw_output'
os.makedirs(output_folder, exist_ok=True)

# Step 2: Open the CSV file for writing
with open(csv_file, 'w', newline='', encoding='utf-8') as f:
    # Define the CSV writer
    writer = csv.writer(f)
    
    # Write the header row
    writer.writerow(['Observation ID', 'Photo ID', 'Photo URL'])
    
    # Step 3-4: Retrieve and download all images associated with each observation
    for observation in observations['results']:
        observation_id = observation['id']
        observation_photos = observation.get('photos', [])
        
        # Download all images associated with the observation
        for photo in observation_photos:
            photo_id = photo['id']
            photo_url = photo['url']
            
            # Download the image and save it to the output folder
            response = requests.get(photo_url)
            image_file_path = f"{output_folder}/photo_{photo_id}.jpg"
            with open(image_file_path, 'wb') as image_file:
                image_file.write(response.content)
            
            # Write metadata to the CSV file
            writer.writerow([observation_id, photo_id, photo_url])
            
            print(f"Downloaded image {photo_id} for observation {observation_id}")

print("All images and metadata downloaded successfully.")
```

```python
import csv
import os
from pyinaturalist import get_observations
import requests

# Define the CSV file path
csv_file = 'Cw_image_metadata.csv'

# Ensure the output folder exists
output_folder = 'Cw_output'
os.makedirs(output_folder, exist_ok=True)

# Open the CSV file for writing
with open(csv_file, 'w', newline='', encoding='utf-8') as f:
    # Define the CSV writer
    writer = csv.writer(f)
    
    # Write the header row
    writer.writerow(['Observation ID', 'Photo ID', 'Photo URL'])
    
    # Specify the initial page number
    page = 1
    
    # Loop to retrieve observations from multiple pages
    while True:
        # Step 1: Search for observations of the species Calycoseris wrightii, specifying per_page=200 and the current page
        observations = get_observations(taxon_name='Calycoseris wrightii', per_page=200, page=page, size = 'large')
        
        # Check if there are no more observations
        if not observations['results']:
            break
        
        # Step 3-4: Retrieve and download all images associated with each observation
        for observation in observations['results']:
            observation_id = observation['id']
            observation_photos = observation.get('photos', [])
            
            # Download all images associated with the observation
            for photo in observation_photos:
                photo_id = photo['id']
                photo_url = photo['url']
                
                # Download the image and save it to the output folder
                response = requests.get(photo_url)
                image_file_path = f"{output_folder}/photo_{photo_id}.jpg"
                with open(image_file_path, 'wb') as image_file:
                    image_file.write(response.content)
                
                # Write metadata to the CSV file
                writer.writerow([observation_id, photo_id, photo_url])
                
                print(f"Downloaded image {photo_id} for observation {observation_id}")
        
        # Increment the page number for the next iteration
        page += 1

print("All images and metadata downloaded successfully.")
```
